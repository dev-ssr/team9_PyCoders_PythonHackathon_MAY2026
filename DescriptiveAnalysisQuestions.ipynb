## Questions for descriptive analysis:

#### 1. Glucose Analysis

1. What is the average glucose level for each patient over time?

Column Chosen:

glucose

Why This Marker Is Important:

Glucose is the primary indicator of diabetes control. Monitoring average glucose levels helps identify patients with poor glycemic control, frequent hyperglycemia, or unstable blood sugar patterns.

2. How often do patients experience hyperglycemia (glucose > 180 mg/dL)?

Column Chosen:

glucose

Why This Marker Is Important:

Frequent hyperglycemia increases the risk of diabetes complications such as kidney disease, nerve damage, and cardiovascular problems.
