# Data Cleaning and Preprocessing

In this notebook, we will perform the initial data cleaning and preprocessing steps for the HUPA-UC Diabetes Dataset. This includes loading the data, merging the different files, handling missing values and inconsistencies, and preparing the data for exploratory data analysis.

```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
master_df = pd.read_csv('merged_data.csv')
```

```python
master_df.shape # checking the number of rows and columns in the dataframe
```

```python
print(master_df['Patient_ID'].unique())
print(len(master_df['Patient_ID'].unique()))
```

```python
master_df['time'] = pd.to_datetime(master_df['time'])
numeric_cols = ['glucose', 'calories', 'heart_rate', 'steps', 'basal_rate', 'bolus_volume_delivered', 'carb_input', 'Age', 'Average Sleep Duration (hrs)', 'Sleep Quality (1-10)', '% with Sleep Disturbances']
```

```python
for col in numeric_cols:
 master_df[col] = pd.to_numeric(master_df[col], errors='coerce')
```

```python
master_df.dtypes
```

## Dropping duplicate rows

```python
master_df.drop_duplicates()
```

```python
master_df.shape
```

Since there is no change in the shape of the dataframe before and after dropping duplicate rows, we can say that there are no duplicate rows in the dataset.

## Handle Missing Values

```python
master_df.isnull().sum()
```

From the above output we can confirm that the dataframe has no Null values.

```python
master_df.dtypes
```

## Format Columns
We can see in the above output that columns glucose, calories and hear_rate have large decimal values. Here we are rounding these column values to 2 decimal places.

```python
master_df['glucose'] = master_df['glucose'].round(2)
master_df['calories'] = master_df['calories'].round(2)
master_df['heart_rate'] = master_df['heart_rate'].round(2)
```

```python
master_df.head()
```

```python
master_df.describe().T
```

From the above output we can seee that bolus_volume_delivered has a negative minimun value. We know that bolus_volume_delivered is a specific, single dose of medication or fluid administered rapidly (typically over 1 to 30 minutes) to achieve a swift therapeutic effect. So this cannot be negative.

## Replacing negative values with 0

```python
for value in master_df['bolus_volume_delivered']:
    if value < 0:
        print(value)
    
```

```python
master_df[master_df['bolus_volume_delivered']<0] = 0
```

```python
master_df.describe().T
```

We can see that the negative values have been replaced by 0.
