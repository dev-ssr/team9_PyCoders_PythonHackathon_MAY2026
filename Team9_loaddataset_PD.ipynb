```python
import pandas as pd
import zipfile
import glob
import os
```

```python
# Path to ZIP file
zip_path = r"C:\Users\Pravallika Dheshidi\OneDrive\Desktop\python hackthon\HUPA-UC Diabetes Dataset.zip"
# Folder to extract files
extract_folder = r"C:\Users\Pravallika Dheshidi\OneDrive\Documents\GitHub\team9_PyCoders_PythonHackathon_MAY2026"
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_folder)
```

```python
# Get all CSV files
csv_files = glob.glob(os.path.join(extract_folder, "*.csv"))
```

```python
print(os.listdir(extract_folder))
```

```python
path = os.path.join(extract_folder, "HUPA-UC Diabetes Dataset", "*.csv")
files = glob.glob(path)

print(files[:5])
```

```python
file_names = [os.path.basename(f) for f in files]
print(file_names[:5])
```

```python
patient_files = [f for f in files if os.path.basename(f).startswith("HUPA")] # just taking files starting with HUPA
print("Patient monitoring files:", len(patient_files))
```

```python
all_data = []

for file in patient_files:
    df = pd.read_csv(file, sep=';')
    
    patient_id = os.path.basename(file).replace(".csv","")
    df["Patient_ID"] = patient_id
    
    all_data.append(df)

master_df = pd.concat(all_data, ignore_index=True)
```

```python
meta_path = os.path.join(
    extract_folder,
    "HUPA-UC Diabetes Dataset",
    "T1DM_patient_sleep_demographics_with_race.csv"
)

meta = pd.read_csv(meta_path)

print(meta.head())
print(meta.columns)
```

```python
master_df = master_df.merge(meta, on="Patient_ID", how="left") # merging sleep demographics with the mater data frame
```
